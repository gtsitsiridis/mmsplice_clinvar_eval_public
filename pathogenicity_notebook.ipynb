```python
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
```

```python
benign_data = np.loadtxt('pathogenicity_benign.csv')
pathogenic_data = np.loadtxt('pathogenicity_pathogenic.csv')
```

```python
fig1, ax1 = plt.subplots()
ax1.set_title('Boxplot - Pathogenicity')
ax1.boxplot([benign_data, pathogenic_data])
ax1.set_xticklabels(['Benign', 'Pathogenic'])
```

```python
fig1, ax1 = plt.subplots()
sm.qqplot(benign_data, line='45', ax=ax1)
ax1.set_xlabel("Theoretical Quantiles")
ax1.set_ylabel("Sample Quantiles")
ax1.set_title('Q-Q Plot - Benign')

fig2, ax2 = plt.subplots()
sm.qqplot(pathogenic_data, line='45', ax=ax2)
ax2.set_xlabel("Theoretical Quantiles")
ax2.set_ylabel("Sample Quantiles")
ax2.set_title('Q-Q Plot - Pathogenic')
```

```python
from scipy.stats import mannwhitneyu

res = mannwhitneyu(benign_data, pathogenic_data, alternative='two-sided')
res.pvalue
```
